```python
df = pd.read_csv('fig_data/Pdiscard_vs_v7/Pdiscard_vs_v7.csv', header=None)
Pdiscard = df.iloc[0].tolist()
```

```python
def main():
    xlist = []
    for i in range(45):
        xlist.append(i*0.02)

    fig = plt.figure(dpi=1200)
    fig_1 = fig.add_subplot(111)
    
    fig_1.plot(xlist, Perror_UW3_9km_5, label="$k=5$")
    fig_1.plot(xlist, Perror_UW3_9km_10, label="$k=10$")
    fig_1.plot(xlist, Perror_UW3_9km_15, label="$k=15$")
    fig_1.plot(xlist, Perror_UW3_9km_20, label="$k=20$")

    #fig_1.plot(xlist, Perror_UW2_9km_5, label="$k=5$")
    #fig_1.plot(xlist, Perror_UW2_9km_10, label="$k=10$")
    #fig_1.plot(xlist, Perror_UW2_9km_15, label="$k=15$")
    #fig_1.plot(xlist, Perror_UW2_9km_20, label="$k=20$")

    fig_2 = fig_1.twinx()
    fig_2.plot(xlist, Pdiscard, label="$k=5$", linestyle='--', color='purple')
    
    fig_1.set_xlabel(r"Discard Window Size Parameter $v_7$")
    fig_1.set_ylabel(r"Error Probability Within a Segment")
    fig_2.set_ylabel(r"Discard Probability")
    fig_1.legend(ncol=2, bbox_to_anchor=(0, 1.1, 1., 0.102), loc=3)
    fig_1.grid(which='minor', alpha=0.2)
    fig_1.grid(which='major', alpha=0.5)
    fig_1.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
    fig_1.ticklabel_format(style="sci",  axis="y",scilimits=(0,0))
    #plt.xticks(xlistgrid)
    plt.xlim(-0.02, 0.8)
    plt.show()

    fig.savefig('FIG9A01.png', bbox_inches="tight", pad_inches=0.05) 
    
if __name__ == "__main__":
    main()
```
