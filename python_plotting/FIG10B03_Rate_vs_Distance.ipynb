```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter
import csv
import pandas as pd
print(pd.__version__)
print(np.__version__)
```

```python
plt.rcParams['font.family'] = 'Times New Roman' 
plt.rcParams['mathtext.fontset'] = 'stix' 
plt.rcParams["font.size"] = 12
plt.rcParams['xtick.labelsize'] = 17
plt.rcParams['ytick.labelsize'] = 17 

plt.rcParams['xtick.direction'] = 'in' 
plt.rcParams['ytick.direction'] = 'in' 
plt.rcParams['axes.grid'] = True 

plt.rcParams["xtick.minor.visible"] = True  
plt.rcParams["ytick.minor.visible"] = True  
plt.rcParams['xtick.top'] = True  
plt.rcParams['ytick.right'] = True  

plt.rcParams["legend.fancybox"] = False 
plt.rcParams["legend.framealpha"] = 1
plt.rcParams["legend.edgecolor"] = 'black'
plt.rcParams["legend.markerscale"] = 5
```

```python
def main():
    
    xlist4 = []
    for i in range(750):
        xlist4.append(4*(i+1))

    xlist5 = []
    for i in range(600):
        xlist5.append(5*(i+1))

    xlist5p5 = []
    for i in range(600):
        xlist5p5.append(5.5*(i+1))

    xlist6 = []
    for i in range(500):
        xlist6.append(6*(i+1))

    xlist7 = []
    for i in range(500):
        xlist7.append(7*(i+1))

    xlist10 = []
    for i in range(300):
        xlist10.append(10*(i+1))

    
    df = pd.read_csv('fig_data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_LP_4km_15.csv', header=None)
    Rate_vs_Distance_LP_k15_4km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_LP_k15_4km = [x*15 for x in Rate_vs_Distance_LP_k15_4km]

    df = pd.read_csv('fig_data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_LP_5km_15.csv', header=None)
    Rate_vs_Distance_LP_k15_5km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_LP_k15_5km = [x*15 for x in Rate_vs_Distance_LP_k15_5km]

    df = pd.read_csv('fig_data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_LP_5p5km_15.csv', header=None)
    Rate_vs_Distance_LP_k15_5p5km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_LP_k15_5p5km = [x*15 for x in Rate_vs_Distance_LP_k15_5p5km]

    df = pd.read_csv('fig_data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_LP_6km_15.csv', header=None)
    Rate_vs_Distance_LP_k15_6km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_LP_k15_6km = [x*15 for x in Rate_vs_Distance_LP_k15_6km]

    df = pd.read_csv('FIG_Data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_LP_7km_15.csv', header=None)
    Rate_vs_Distance_LP_k15_7km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_LP_k15_7km = [x*15 for x in Rate_vs_Distance_LP_k15_7km]

    df = pd.read_csv('fig_data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_LP_Spool_125m_4km_15.csv', header=None)
    Rate_vs_Distance_LP_Spool_125m_k15_4km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_LP_Spool_125m_k15_4km = [x*15 for x in Rate_vs_Distance_LP_Spool_125m_k15_4km]

    df = pd.read_csv('fig_data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_LP_Spool_125m_5km_15.csv', header=None)
    Rate_vs_Distance_LP_Spool_125m_k15_5km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_LP_Spool_125m_k15_5km = [x*15 for x in Rate_vs_Distance_LP_Spool_125m_k15_5km]

    df = pd.read_csv('fig_data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_LP_Spool_125m_5p5km_15.csv', header=None)
    Rate_vs_Distance_LP_Spool_125m_k15_5p5km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_LP_Spool_125m_k15_5p5km = [x*15 for x in Rate_vs_Distance_LP_Spool_125m_k15_5p5km]

    df = pd.read_csv('fig_data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_LP_Spool_125m_6km_15.csv', header=None)
    Rate_vs_Distance_LP_Spool_125m_k15_6km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_LP_Spool_125m_k15_6km = [x*15 for x in Rate_vs_Distance_LP_Spool_125m_k15_6km]

    df = pd.read_csv('FIG_Data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_LP_Spool_125m_7km_15.csv', header=None)
    Rate_vs_Distance_LP_Spool_125m_k15_7km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_LP_Spool_125m_k15_7km = [x*15 for x in Rate_vs_Distance_LP_Spool_125m_k15_7km]

    #PLOB bound
    eta_list = [10**(-L/22) for L in xlist10]
    PLOB_rate = [-np.log2(1 - eta) if eta < 1 else 0 for eta in eta_list]
    
    fig = plt.figure(dpi=1200)
    fig_1 = fig.add_subplot(111)

    fig_1.plot(xlist4, Total_Rate_vs_Distance_LP_k15_4km, label="$k=15$, $L_0=4km$", color="orange", linewidth="2")
    fig_1.plot(xlist5, Total_Rate_vs_Distance_LP_k15_5km, label="$k=15$, $L_0=5km$", color="#e377c2", linewidth="2")
    fig_1.plot(xlist5p5, Total_Rate_vs_Distance_LP_k15_5p5km, label="$k=15$, $L_0=5.5km$", color="r", linewidth="2")
    fig_1.plot(xlist6, Total_Rate_vs_Distance_LP_k15_6km, label="$k=15$, $L_0=6km$", color="#1f77b4", linewidth="2")
    fig_1.plot(xlist7, Total_Rate_vs_Distance_LP_k15_7km, label="$k=15$, $L_0=7km$", color="g", linewidth="2")

    fig_1.plot(xlist10, PLOB_rate, label="PLOB Bound", color="blue", linestyle="--")

    fig_1.plot(xlist4, Total_Rate_vs_Distance_LP_Spool_125m_k15_4km, label=r"$k=15$, $L_0=4km$, $L_{\mathrm{TEC}}=125m$", color="orange", linestyle="--", linewidth="2")
    fig_1.plot(xlist5, Total_Rate_vs_Distance_LP_Spool_125m_k15_5km, label=r"$k=15$, $L_0=5km$, $L_{\mathrm{TEC}}=125m$", color="#e377c2", linestyle="--", linewidth="2")
    fig_1.plot(xlist5p5, Total_Rate_vs_Distance_LP_Spool_125m_k15_5p5km, label=r"$k=15$, $L_0=5.5km$, $L_{\mathrm{TEC}}=125m$", color="r", linestyle="--", linewidth="2")
    fig_1.plot(xlist6, Total_Rate_vs_Distance_LP_Spool_125m_k15_6km, label=r"$k=15$, $L_0=6km$, $L_{\mathrm{TEC}}=125m$", color="#1f77b4",linestyle="--", linewidth="2")
    fig_1.plot(xlist7, Total_Rate_vs_Distance_LP_Spool_125m_k15_7km, label=r"$k=15$, $L_0=7km$, $L_{\mathrm{TEC}}=125m$", color="g", linestyle="--", linewidth="2")

        
    fig_1.set_xlabel(r"$L_{\mathrm{total}}$ [km]")
    fig_1.set_ylabel(r"$R_{\mathrm{protocol\ run}}$ [ebit/protocol run]")
    fig_1.legend(ncol=2, bbox_to_anchor=(0, 1.1, 1., 0.102), loc=3)
    fig_1.grid(which='minor', alpha=0.2)
    fig_1.grid(which='major', alpha=0.5)
    fig_1.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
    #fig_1.ticklabel_format(style="sci",  axis="y",scilimits=(0,0))
    plt.xlim(0, 3000)
    plt.ylim(1e-4, 20)
    plt.yscale("log")
    plt.show()

    fig.savefig('FIG10B03.png', bbox_inches="tight", pad_inches=0.05) 

if __name__ == "__main__":
    main()
```
